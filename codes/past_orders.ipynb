```python
import pandas as pd
import numpy as np

import datetime
from datetime import timedelta
import random
from random import randint
```

```python
random.seed(10)
```

```python
customers_df = pd.read_csv('../data/customers.csv')
addresses_df = pd.read_csv('../data/addresses.csv')
```

```python
# generate phone number
for i in range (customers_df.shape[0]):
    rand_tail = randint(1000000000, 9999999999)
    rand_phone = '+44' + str(rand_tail)
    customers_df.at[i,'phone_number'] = rand_phone
```

```python
# create output df
col_names = ['order_id', 'customer_id', 'order_date', 'product_id', 'quantity', 'customer_name', 'shipping_address', 'phone_number', 'price', 'partner']
df = pd.DataFrame(0, index=np.arange(100), columns=col_names)
```

```python
# generate random customer_id
customer_num_list = []
for i in range(5): # make range 5 to have at least 5 orders per customer
    customer_num_list += list(range(20))


random.shuffle(customer_num_list)

# update customer_id
for i in range(len(customer_num_list)):
    customer_id = customers_df.loc[customer_num_list[i], 'ID']
    df.at[i, 'customer_id'] = customer_id

    customer_name = customers_df.loc[customer_num_list[i], 'Customers']
    df.at[i, 'customer_name'] = customer_name

    shipping_address = customers_df.loc[customer_num_list[i], random.choice(['Address1', 'Address2'])]
    df.at[i, 'shipping_address'] = shipping_address

    phone_number = customers_df.loc[customer_num_list[i], 'phone_number']
    df.at[i, 'phone_number'] = phone_number
```

```python
# generate random date
date_list = []

start_date = datetime.date(2022, 10, 1)
end_date   = datetime.date(2022, 11, 30)

for i in range(df.shape[0]):
    random_date = start_date + (end_date - start_date) * random.random()
    date_list.append(random_date)
date_list.sort()

# update date
for i in range(len(date_list)):
    df.at[i,'order_date'] = date_list[i]
```

## cost, delivery fee, total cost, partner

```python
books_df = pd.read_csv('../data/books.csv')
cd_df = pd.read_csv('../data/cd.csv')
fresh_df = pd.read_csv('../data/fresh.csv')
home_appliance_df = pd.read_csv('../data/home_appliance.csv')
phone_df = pd.read_csv('../data/phone.csv')
partners_df = pd.read_csv('../data/partners.csv')
```

```python
# create output df
col_names = ['product_id', 'price']
product_df = pd.DataFrame(books_df.iloc[:,[0,7]])
product_df.columns = col_names
```

```python
# cd
cd_df = cd_df.rename(columns={'price (£)':'price'})
product_df = pd.concat([product_df, cd_df.iloc[:,[0,2]]])
```

```python
# fresh
fresh_df = fresh_df.rename(columns={'standard_price_to_customers':'price'})
product_df = pd.concat([product_df, fresh_df.iloc[:,[0,9]]])
```

```python
# home appliance
product_df = pd.concat([product_df, home_appliance_df.iloc[:,[0,2]]])
```

```python
# phone
product_df = pd.concat([product_df, phone_df.iloc[:,[0,2]]])
```

```python
product_df.head()
```

```python
product_df.index = list(range(0,product_df.shape[0]))
```

```python
for i in range(df.shape[0]):
    #order_id
    order_id = i + 1
    df.at[i, 'order_id'] = order_id

    # quantity
    quantity = random.randint(1, 5)
    df.at[i, 'quantity'] = quantity

    # product_id
    product_num = randint(0,product_df.shape[0]-1)
    df.at[i, 'product_id'] = product_df.loc[product_num, 'product_id']

    # price
    df.at[i, 'price'] = product_df.loc[product_num, 'price']
    # partner
    partner_num = randint(0,partners_df.shape[0]-1)
    df.at[i, 'partner'] = partners_df.loc[partner_num, 'Partner_id']
```
